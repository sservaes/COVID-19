```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

# Background from data
Data is coming from https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases/resource/00fa0e37-961b-4767-a5ce-e7ab4e2c921c

```python
covid_df = pd.read_csv("time_series_covid19_confirmed_global.csv")
covid_df.head()
```

```python
worst = covid_df[['Country/Region', '3/26/20']]
worst.head()
worst = worst.sort_values(by = worst.columns[1], axis = 0, ascending = False)
worst.head()
plt.figure(figsize = (10,4))
plt.bar('Country/Region', worst.columns[1], data = worst[0:9])
plt.xticks(rotation=90)
plt.title(worst.columns[1])
```
